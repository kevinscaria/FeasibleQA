```python
import pandas as pd
```

```python
def getResClean(df):
  # Stage 1
  df['GPT2PredClean1'] = df['GPT2Pred'].apply(lambda x: x.split('Result:')[-1].split('\n')[0].strip())

  # Deep clean
  allGPTCleanedOutput = []
  for idx, row in enumerate(df['GPT2PredClean1']):
    optList = []
    if '{' in row and row.count('{')!=row.count('}'):
      row = row[1:]
      rec = row.split(', ')[:-1]
      for i in rec:
        optList.append(i.split(':')[-1].lstrip().strip('}'))
        optList = list(set(optList))
    elif '{' in row and row.count('{')==row.count('}'):
      optList = []
      row = row.lstrip('[').rstrip(']').rstrip(',')
      rec = row.split(', ')
      for i in rec:
        optList.append(i.split(':')[-1].lstrip().strip('}').replace("'", ''))
        optList = list(set(optList))
    else:
      optList = row
    allGPTCleanedOutput.append(optList)
  return allGPTCleanedOutput
```

```python
df = pd.read_csv('./GPT2/MCQ/FewShot_WK_MCQ_Output.csv')
display(df.head())

df['GPT2PredClean2'] = getResClean(df)
df[['correct_answers', 'GPT2PredClean2']].apply(lambda x: x[0] == x[1], axis=1).sum()*100/len(df), (df[['correct_answers', 'GPT2PredClean2']].apply(lambda x: x[0] == x[1], axis=1).sum(), len(df))
```
